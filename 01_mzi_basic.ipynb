```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
```

```python
#No phase (all goes to |0⟩)
qc1 = QuantumCircuit(1, 1)
qc1.h(0)
qc1.rz(0, 0)        # no phase
qc1.h(0)
qc1.measure(0, 0)

simulator = AerSimulator()
res1 = simulator.run(qc1, shots=1000).result().get_counts()
print("No phase:", res1)

fig = plot_histogram(res1)
fig.suptitle("MZI — No Phase (θ = 0)")
fig.savefig("mzi_no_phase.png", dpi=150, bbox_inches='tight')
display(fig)
```

```python
qc2 = QuantumCircuit(1, 1)
qc2.h(0)
qc2.rz(np.pi, 0)
qc2.h(0)
qc2.measure(0, 0)

simulator = AerSimulator()
res2 = simulator.run(qc2, shots=1000).result().get_counts()
print("Pi phase:", res2)

fig = plot_histogram(res2)
fig.suptitle("MZI — Full Phase (θ = π)")
fig.savefig("mzi_pi_phase.png", dpi=150, bbox_inches='tight')
display(fig)
```

```python
qc3 = QuantumCircuit(1, 1)
qc3.h(0)
qc3.rz(np.pi/2, 0)
qc3.h(0)
qc3.measure(0, 0)

simulator = AerSimulator()
res3 = simulator.run(qc3, shots=1000).result().get_counts()
print("Pi/2 phase:", res3)

fig = plot_histogram(res3)
fig.suptitle("MZI — Partial Phase (θ = π/2)")
fig.savefig("mzi_pi2_phase.png", dpi=150, bbox_inches='tight')
display(fig)
```
